# **STATEMENT OF PURPOSE**

Through this project we aim to leverage natural language toolkit to perform a **sentiment analysis** on a dataset that consists of reviews left by customers for the AirBnBs they stayed in. The goal is to understand the customer's perspective, which will influence us to drive change in the way we approach our customer service strategy.

## **Contents**

1. Tokenization through NLTK
2. Bar Chart
3. Word Cloud
4. Frequency Bar charts for types of words
5. Sentiment Analysis
6. Top 10 Positive/Negative words by frequency.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
import matplotlib.pyplot as plt
```

```python
nltk.download('stopwords')
nltk.download('punkt')
```

```python
# Replace 'your_reviews_data.txt' with the actual file path or data loading method
with open('/content/reviews.csv', 'r', encoding='utf-8') as file:
    reviews_data = file.read()
```

```python
# Tokenize the text into words
words = word_tokenize(reviews_data.lower())  # Convert to lowercase for consistency

# Remove stopwords and non-alphabetic words
filtered_words = [word for word in words if word.isalpha() and word not in stopwords.words('english')]
```

```python
positive_words = ['good', 'excellent', 'amazing', 'outstanding', 'fantastic', 'beautiful', 'great', 'wonderful', 'comfortable']
negative_words = ['bad', 'poor', 'terrible', 'horrible', 'awful', 'discusting', 'not friendly', 'unfriendly', 'messy', 'uncomfortable', 'not happy' 'not good', 'not clean' ]

positive_count = sum(word in positive_words for word in filtered_words)
negative_count = sum(word in negative_words for word in filtered_words)

print(f"Positive words count: {positive_count}")
print(f"Negative words count: {negative_count}")
```

```python
# Create a bar chart
labels = ['Positive', 'Negative']
counts = [positive_count, negative_count]

plt.bar(labels, counts, color=['green', 'red'])
plt.title('Sentiment Analysis of Airbnb Reviews')
plt.xlabel('Sentiment')
plt.ylabel('Word Count')
plt.show()
```

```python
from wordcloud import WordCloud
```

```python
# Combine positive and negative words
all_words = positive_words + negative_words + filtered_words

# Create a word frequency distribution
word_freq = FreqDist(all_words)

# Generate a word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)

# Display the word cloud using Matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Airbnb Reviews')
plt.show()
```

```python
# Count occurrences of positive and negative words
positive_word_counts = {word: filtered_words.count(word) for word in positive_words}
negative_word_counts = {word: filtered_words.count(word) for word in negative_words}

# Visualize the results
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Bar plot for positive words
ax1.bar(positive_word_counts.keys(), positive_word_counts.values(), color='green')
ax1.set_title('Positive Words Frequency')
ax1.set_xlabel('Words')
ax1.set_ylabel('Frequency')

# Bar plot for negative words
ax2.bar(negative_word_counts.keys(), negative_word_counts.values(), color='red')
ax2.set_title('Negative Words Frequency')
ax2.set_xlabel('Words')
ax2.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from nltk.corpus import opinion_lexicon
import matplotlib.pyplot as plt

# Download NLTK resources
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('opinion_lexicon')

# Load your Airbnb reviews data
with open('/content/reviews.csv', 'r', encoding='utf-8') as file:
    reviews_data = file.read()

# Tokenize the text into words
words = word_tokenize(reviews_data.lower())

# Remove stopwords and non-alphabetic words
filtered_words = [word for word in words if word.isalpha() and word not in stopwords.words('english')]

# Calculate word frequencies
freq_dist = FreqDist(filtered_words)

# Set a threshold for word frequency (adjust as needed)
threshold = 10

# Get words that occur more frequently than the threshold
common_words = [word for word, count in freq_dist.items() if count > threshold]

# Use NLTK's opinion lexicon to identify negative words
negative_lexicon = set(opinion_lexicon.negative())

# Filter common words to get potential negative words
potential_negative_words = [word for word in common_words if word in negative_lexicon]

# Display potential negative words
print("Potential Negative Words:", potential_negative_words)
```

```python
# Use NLTK's opinion lexicon to identify negative words
positive_lexicon = set(opinion_lexicon.positive())

# Filter common words to get potential negative words
potential_positive_words = [word for word in common_words if word in positive_lexicon]

# Display potential negative words
print("Potential Positive Words:", potential_positive_words)
```
